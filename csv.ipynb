```python
import serial
import time

serial_port = "COM12"  # Change this to the correct port
baud_rate = 9600

# Open the serial port
with serial.Serial(serial_port, baud_rate) as ser:
    # Open the CSV file in append mode
    with open("roomdata.csv", "a", buffering=1) as file:
        # Write the header only if the file is empty
        if file.tell() == 0:
            file.write("Room Occupancy, People, Temperature, Humidity, Time Occupied\n")
        
        while True:
            # Read data from the serial port and clean it
            data = ser.readline().decode("utf-8", errors="ignore").strip()
            
            if data:
                # Try to parse the data
                try:
                    if "Room Occupancy:" in data and "People:" in data and "Temp:" in data and "Humidity:" in data and "Time Occupied:" in data:
                        # Split the decoded string into key-value pairs
                        key_value_pairs = data.split(',')

                        # Initialize variables
                        room_occupancy = None
                        people = temp = humidity = time_occupied = None

                        # Process each key:value pair
                        for pair in key_value_pairs:
                            key, value = pair.split(':')  # Split by ':'
                            key = key.strip()  # Remove any leading/trailing whitespace
                            value = value.strip()  # Remove any leading/trailing whitespace

                            # Assign values to the corresponding variables
                            if key == "Room Occupancy":
                                room_occupancy = 1 if value.lower() == "occupied" else 0
                            elif key == "People":
                                people = int(value)
                            elif key == "Temp":
                                temp = float(value)
                            elif key == "Humidity":
                                humidity = float(value)
                            elif key == "Time Occupied":
                                time_occupied = int(value)

                        # Print and write the data to the CSV file
                        print(f"Room Occupancy: {room_occupancy}, People: {people}, Temp: {temp}, Humidity: {humidity}, Time Occupied: {time_occupied}")
                        file.write(f"{room_occupancy}, {people}, {temp}, {humidity}, {time_occupied}\n")
                        file.flush()
                        time.sleep(2)
                except (IndexError, ValueError) as e:
                    print(f"Error parsing data: {data} - {e}")
```
